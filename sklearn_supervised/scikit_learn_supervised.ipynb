```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
# load Boston dataset to boston
boston = load_boston()

print(f'Boston dataset\'s keys: {boston.keys()}')
```

```python
#load feature names
feature_names = boston["feature_names"]

print(f'Boston dataset\'s feature names: {feature_names}')
```

```python
# load Boston's DS data
data = boston["data"]

data.shape
```

```python
print(boston["DESCR"])
```

```python
#load target array
target = boston["target"]
print(f'First 5 targets: {target[:5]}')
```

```python
# Create X dataframe with common data
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
# Information about X dataframe
X.info()
```

```python
# Create y dataframe with target data
y = pd.DataFrame(target, columns=['price'])
y.head()
```

```python
# Information about y dataframe
y.info()
```

##### Split dataframes X & y on train & test ones

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

##### Create linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

#### Teach model & test it

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred.shape
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(15)
```

#### Calculate R2 score

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
# write dataframe 'df' without column Class to dataframe 'X'
X = pd.DataFrame(df.loc[:, df.columns != 'Class'])
```

```python
X.head()
```

```python
# write column Class from dataframe 'df' to Series 'y'
y = pd.Series(df['Class'])
```

```python
# split dataframes X & y on train & test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
# show shapes of train and test datasets
print(f'The shape of the dataframe X_train is {X_train.shape}.')
print(f'The shape of the dataframe X_test is {X_test.shape}.')
print(f'The shape of the dataframe y_train is {y_train.shape}.')
print(f'The shape of the dataframe y_test is {y_test.shape}.')
```

Create GridSearchCV model.

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]

clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)
```

```python
# train a model
clf.fit(X_train, y_train)
```

```python
# show the best parameters for this model
print(f'The best parameters for the model: {clf.best_params_}')
```

Calculate AUC

```python
# use methos .predict_proba
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba
```

```python
# write column 1 to y_pred_proba
y_pred_proba = y_pred_proba[:, 1]
```

```python
print(y_pred_proba[:5])
```

##### Calculate AUC

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
